```python
import os
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
from langchain_groq import ChatGroq
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_openai import OpenAIEmbeddings
```

```python
llm = ChatGroq(
    model="qwen/qwen3-32b"
)
```

```python
# embedding_model = GoogleGenerativeAIEmbeddings(
#     model="models/embedding-001"
# )

embedding_model = OpenAIEmbeddings(
    model="text-embedding-3-small",
    api_key=os.getenv("OPENAI_API_KEY")
)
```

# 1. Data Ingestion

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS

from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.documents import Document
from langchain_core.runnables import RunnablePassthrough
```

```python
os.getcwd()
```

```python
os.path.join(os.getcwd(), "data", "sample.pdf")
```

```python
loader = PyPDFLoader(file_path=os.path.join(os.getcwd(), "data", "sample.pdf"))
```

```python
documents = loader.load()
```

```python
len(documents)
```

```python
print(documents[0].page_content)
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=150,
    length_function=len
)
```

```python
chunks = text_splitter.split_documents(documents=documents)
```

```python
len(chunks)
```

```python
print(chunks[0].page_content)
```

```python
chunks[0].metadata
```

```python
chunks[1].metadata
```

```python
vector_store = FAISS.from_documents(
    documents=chunks,
    embedding=embedding_model
)
```

```python
vector_store.similarity_search(query="llama2 finetuning benchmark experiments.", k=3)
```

```python
retriever = vector_store.as_retriever(search_kwargs={'k':4})
```

```python
retriever.invoke(input="llama2 finetuning benchmark experiments.")
```

```python
prompt_template = """
        Answer the question based on the context provided below. 
        If the context does not contain sufficient information, respond with: 
        "I do not have enough information about this."

        Context: {context}

        Question: {question}

        Answer:"""
```

```python
prompt = PromptTemplate(
    template=prompt_template,
    input_variables=['context', 'question']
)
```

```python
prompt.pretty_print()
```

```python
prompt.invoke(
    input={
        'context': "Here is the dummy context.",
        'question': 'Here is the dummy queestion'
    }
)
```

```python
def format_documents(documents):
    return "\n\n".join(doc.page_content.strip() for doc in documents if (isinstance(doc, Document) and doc.page_content!=""))
```

```python
rag_chain = (
    {
        'context': retriever | format_documents,
        'question': RunnablePassthrough()
    }
    |prompt
    |llm
    |StrOutputParser()
)
```

```python
response = rag_chain.invoke("tell  me about the llama2 finetuning benchmark experiments?")
```

```python
print(response)
```

```python
response = rag_chain.invoke("can you tell me Scaling trends for the reward model?")
```
