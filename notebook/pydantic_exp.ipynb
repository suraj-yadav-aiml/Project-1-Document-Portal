```python
from pydantic import RootModel
from typing import List
```

```python
Pets = RootModel[list[str]]
```

```python
p = Pets(['dog','cat'])
```

```python
p.root
```

```python
p.model_dump()
```

```python
p.model_dump_json()
```

```python
class MyPets(RootModel):
    root: List[str]
```

```python
p = MyPets.model_validate(['dog','cat'])
```

```python
p.root
```

```python
p.model_dump()
```

```python
p.model_dump_json()
```
